```python
df = df.drop(columns=["Unnamed: 19"])
```

```python
df = df.drop(columns=["Density conversion factor"])
```

```python
df = df.rename(columns={"Inudstry Type": "Industry Type"})
```

```python
df["APCD Type"] = df["APCD Type"].fillna("Unknown")
```

```python
print(df["Avg hours/day: "].unique())
print(df["Load factor - non DG set (%):"].unique())
```

```python
df["Avg hours/day: "] = pd.to_numeric(
    df["Avg hours/day: "], errors="coerce"
)

df["Load factor - non DG set (%):"] = pd.to_numeric(
    df["Load factor - non DG set (%):"], errors="coerce"
)
```

```python
# Numeric columns
numeric_cols = [
    "Rated capacity",
    "Avg hours/day: ",
    "Days/Year",
    "Load factor - non DG set (%):",
    "Avg Fuel used per day:(kg/kl/kw)",
    "VOC(kg/day)"
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Categorical columns
categorical_cols = [
    "Fuel Type",
    "Capacity Unit (TPH / MW / kVA / (cal/hr))",
    "Fuel unit per day (kg/kl)",
    "APCD Type",
    "Any CEMS installed? Yes/No. "
]

for col in categorical_cols:
    if col == "APCD Type":
        df[col] = df[col].fillna("Unknown")
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
X = df[
    [
        "Industry Type",
        "Equipment",
        "Fuel Type",
        "Any CEMS installed? Yes/No.",
        "Rated capacity",
        "Capacity Unit (TPH / MW / kVA / (cal/hr))",
        "Avg hours/day:",
        "Days/Year",
        "Load factor - non DG set (%):",
        "Avg Fuel used per day:(kg/kl/kw)",
        "Fuel unit per day (kg/kl)",
        "APCD Type"
    ]
]

Y = df[
    [
        "PM10(kg/day)",
        "PM2.5(kg/day)",
        "SO2(kg/day)",
        "NO2(kg/day)",
        "CO(kg/day)",
        "VOC(kg/day)"
    ]
]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
categorical_features = [
    "Industry Type",
    "Equipment",
    "Fuel Type",
    "Any CEMS installed? Yes/No.",
    "Capacity Unit (TPH / MW / kVA / (cal/hr))",
    "Fuel unit per day (kg/kl)",
    "APCD Type"
]

numeric_features = [
    "Rated capacity",
    "Avg hours/day:",
    "Days/Year",
    "Load factor - non DG set (%):",
    "Avg Fuel used per day:(kg/kl/kw)"
]
```

```python
print(X.shape)
print(Y.shape)

print(X_train.shape)
print(X_test.shape)

print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        (
            "categorical",
            OneHotEncoder(handle_unknown="ignore"),
            categorical_features
        )
    ],
    remainder="passthrough"
)
```

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import RandomForestRegressor

model = MultiOutputRegressor(
    RandomForestRegressor(
        n_estimators=200,
        random_state=42
    )
)
```

```python
pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ]
)
```

```python
pipeline.fit(X_train, Y_train)
```

```python
predictions = pipeline.predict(X_test)
```

```python
predictions[:5]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(Y_test, predictions)
rmse = mean_squared_error(Y_test, predictions) ** 0.5
r2 = r2_score(Y_test, predictions)

print("MAE :", mae)
print("RMSE:", rmse)
print("R²  :", r2)
```
